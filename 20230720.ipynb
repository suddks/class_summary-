- 인덱스로 리스트 순회
    
    리스트의 요소가 아닌 인덱스로 접근하여 해당요소들을 변경하기

```python
numbers = [4, 6, 10, -8, 5]

for i in range(len(numbers)):   #len()쓰는 이유 : 재사용성 위해 사용
    numbers[i] = numbers[i] * 2

print(numbers)
```

```python
# 이중 반복문

outers = ['A', 'B']
inners = ['c', 'd']

for outer in outers:
    for inner in inners:
        print(outer, inner)   #안쪽 반복문이 다 끝나야 위로 올라갈 수 있음
 
```

```python
# 중첩 리스트 순회

elements = [['A', 'B'], ['c', 'd']]

for items in elements:
    for items2 in items:
        print(items2)
```

### while 

    조건이 참인 동안 코드를 반복해서 실행 = 조건이 거짓이 될 때 까지 반복

    * 종료조건에 다가갈 수 있도록 해야함 => 무한 루프.. 

    while 조건식:
        코드블록


```python
number = int(input('양의 정수를 입력해 주세요 : '))

while number <= 0:
    if number < 0:
        print('음수를 입력했습니다.')
    else:
        print('0은 양의 정수가 아닙니다.')

    number = int(input("양의 정수를 입력해주세요 : "))

print("잘했습니다!")
```

- break / contiune

    - break : 반복을 즉시 중지

```python
number = int(input('양의 정수를 입력해 주세요 : '))

while number <= 0:
    if number == -9999:
        break
    if number < 0:
        print('음수를 입력했습니다.')
    else:
        print('0은 양의 정수가 아닙니다.')

    number = int(input("양의 정수를 입력해주세요 : "))

print("잘했습니다!")
```

## list comprehension

    새로운 list를 만드는 것이 목적 

    간결하고 효율적인 리스트 생성 방법

    가독성이 떨어짐 

    - [expression for 변수 in iterable if 조건식]

    - list(expression for 변수 in iterable if 조건식)

```python
# 0~9 요소를 가지는 리스트 만들기 

new_list = []
for i in range(10):
    # if i % 2 == 1:
        new_list.append(i)
print(new_list)

# list comprehension 적용
new_list_2 = [i for i in range(10)] # if i % 2 == 1]
new_list_3 = [i if i % 2 == 1 else str(i) for i in range(10)]
print(new_list_2)
print(new_list_3)
```

- 리스트를 생성하는 3가지 방법 비교 

    1. for 반복문
    2. map
    3. [expression for 변수 in iterable if 조건식]


```python
# 정수 1, 2, 3을 가지는 새로운 리스트 만들기
numbers = ['1', '2', '3']

# for 반복문
new_numbers = []
for number in numbers:
    new_numbers.append(int(number))
print(new_numbers)

# map 함수

new_numbers_2 = list(map(int,numbers))
print(new_numbers_2)

#list expression 

new_numbers_3 = [int(number) for number in numbers]
print(new_numbers_3)
```

- enumerate : 리스트에서 인덱스랑 값을 같이 가지고 오고 싶을 때 


```python
fruits = ['apple', 'banana', 'cherry']

for index, fruit in enumerate(fruits):
    print(f'인덱스 {index}: {fruit}')
```

```python
result = ['a', 'b','c']
print(list((enumerate(result)))) 
  #값의 요소가 두개 -> 인덱스 + 값

for index, elem in enumerate(result):
    print(index, elem) 
# emnumerate 한 것이 튜플이기 때문에 index, elem 에 각각 들어간다
```

```python
for i in range(40):
    pass
else:
    pass

# else문은 반복문을 중간 종료 없이 실행한 이후에 실행 
# 반복문이 break 문으로 종료될떄는 else문이 실행되지 않는다.
# break 를 통해 중간에 종료되지 않은 경우만 else문이 실행

# 반복문이 중간에 종료된 적이 있는지를 검사할 때 유용하게 사용


string1 = "apple"

for c in string1:
    #apple 문자열을 순회하면서 알파벳 b 가 있으면 b!를 출력한다음 반복문 종료
    if c == 'b':
        print("b!")
        break
else : 
    print("이 단어 안에는 b가 없습니다.")
```
